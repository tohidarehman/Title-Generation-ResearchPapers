```python
# function to return a list of matched entities between abstract and source article
import re
def entityMatch(ent, source):
  ent_split = ent.split()
  result = []
  for l in range(len(ent_split), 1, -1):
    for start_i in range(len(ent_split) - l + 1):
      sub_ent = " ".join(ent_split[start_i:start_i+l])
      if re.search(re.escape(sub_ent), source, re.IGNORECASE):
        result.append(sub_ent)
    if result:
      break
  if result:
    return result
  else:
    for token in ent_split:
      if token.lower() not in STOP_WORDS:
        if re.search(re.escape(token), source, re.IGNORECASE):
          result.append(token)
    return result
  #print([])
  return []
```

```python
#finding unique list
def unique1(list1):
    unique_list = pd.Series(list1).drop_duplicates().tolist()
    return unique_list
    #for x in unique_list:
        #print(x)
```

```python
# function to calculate precision-source metric
def calcPrecS(predicted_abstract,sourcetext):
    abs = nlp(predicted_abstract)
    art = nlp(sourcetext)
    #abs = list(abs.sents)
    matched_entities = []
    entities_per_abs = []
    entities_per_article=[]
    for ee in art.ents:
      entities_per_article.append(ee.text)
    for e in abs.ents:
      entities_per_abs.append(e.text)
      #print(entities_per_abs)
      match_result = entityMatch(e.text, sourcetext)
      if match_result:
        matched_entities.append(match_result[0])
        #print(matched_entities)
    #print("article",len(entities_per_article))
    #print("predicted",len(entities_per_abs))
    #print(len(matched_entities))


    #without using unique set
    entities_per_abs=entities_per_abs
    entities_per_article=entities_per_article
    matched_entities=matched_entities

    #print(len(matched_entities))

    if not entities_per_abs:
      precS=0
    elif not entities_per_abs and not entities_per_article:
      precS=1
    else:
      precS = len(matched_entities)*1.00/len(entities_per_abs)
    return precS #, entities_per_article,entities_per_abs,matched_entities]
```

```python
# function to calculate precision-target and recall-target and F1-target metrics
def calcPrecT(predicted_abstract,  abstract ):
    generated_summ = nlp(predicted_abstract)
    abs = nlp(abstract)
    matched_entities = []
    entities_per_predicAbs = []
    entities_per_abs = []
    for e in abs.ents:
      entities_per_abs.append(e.text)
    for e in generated_summ.ents:
      entities_per_predicAbs.append(e.text)
      match_result = entityMatch(e.text, abstract)
      if match_result:
        matched_entities.append(match_result[0])

    #print(entities_per_predicAbs)
    #print(matched_entities)
    #print(entities_per_abs)


    matched_entities=matched_entities
    entities_per_predicAbs=entities_per_predicAbs
    entities_per_abs=entities_per_abs
    #calculating precision of target
    if not entities_per_predicAbs:
      precT=0
    elif not entities_per_abs and not entities_per_predicAbs:
      precT=1
    else:
      precT = len(matched_entities)*1.00/len(entities_per_predicAbs)

      #calculating recall of target
    if not entities_per_abs:
      recallT=0
    elif not entities_per_abs and not entities_per_predicAbs:
      recallT=1
    else:
      recallT = len(matched_entities)*1.00/len(entities_per_abs)
      #calculating F1 of target
    if (precT==0 and recallT==0):
      F1T=0
    else:
      F1T= 2*precT*recallT*1.00/(precT+recallT)

    return [precT, recallT, F1T, entities_per_abs,entities_per_predicAbs,matched_entities]
```
